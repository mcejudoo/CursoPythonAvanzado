### Obtener los 5 nombres más utilizados en la década de los 90 en EEUU

```python
ini = 1990
fin = 1999
```

```python
L = os.listdir(path)
L[:5]
```

```python
def filtro(fich):
    aux = fich[3:7]  
    if aux.isnumeric():
        año = int(aux)
        return ini <= año <= fin
    else:
        return False
```

```python
L2 = [f for f in L if filtro(f)]
L2
```

```python
primero = True
for fich in L2:
    path_fich  = path + fich
    dtYY = pd.read_csv(path_fich, header=None, names=['nombre','sexo','cuenta'])
    dtYY.set_index(['nombre','sexo'], inplace=True)
    
    if primero:
        total = dtYY
        primero = False
    
    else:
        total = total.add(dtYY, fill_value=0)    
```

```python
total.sort_values(by='cuenta', ascending=False, inplace=True)
total.head()
```
